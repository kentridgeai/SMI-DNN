```python
import pickle
import numpy as np
from collections import defaultdict

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
mpl.rcParams["mathtext.fontset"] = 'cm'
```

### Plots for Section 2.4: SMI Behaviour in DNNs

#### 1. MLP, MNIST

```python
cfg = {}
cfg['dataset'] = 'MNIST'
cfg['model'] = 'MLP'
cfg['width'] = [1024,1024,1024,1024]
cfg['optimizer'] = 'SGD'
arch =  '-'.join(map(str,cfg['width']))
exp_name = 'dpi/'+cfg['model']+'_'+arch+'_'+cfg['dataset']+'_'+cfg['optimizer']

fname = exp_name+'/smi_0'
with open(fname, 'rb') as f:
    smi_initial = pickle.load(f)

smi_initial = np.array(smi_initial)
mean_smi_initial = np.mean(smi_initial, axis=0)
min_smi_initial = np.min(smi_initial, axis=0)
max_smi_initial = np.max(smi_initial, axis=0)

fname = exp_name+'/smi_5'
with open(fname, 'rb') as f:
    smi_mid = pickle.load(f)

smi_mid = np.array(smi_mid)
mean_smi_mid = np.mean(smi_mid, axis=0)
min_smi_mid = np.min(smi_mid, axis=0)
max_smi_mid = np.max(smi_mid, axis=0)

fname = exp_name+'/smi_50'
with open(fname, 'rb') as f:
    smi_final = pickle.load(f)

smi_final = np.array(smi_final)
mean_smi_final = np.mean(smi_final, axis=0)
min_smi_final = np.min(smi_final, axis=0)
max_smi_final = np.max(smi_final, axis=0)
```

```python
plt.figure(figsize=(4,3),dpi=100)
plt.plot(range(6), mean_smi_initial, c='blue', marker='o', label='Epoch 0')
plt.fill_between(range(6), min_smi_initial, max_smi_initial, alpha=0.2, color='tab:blue')
plt.plot(range(6), mean_smi_mid, c='red', marker='x', label='Epoch 5')
plt.fill_between(range(6), min_smi_mid, max_smi_mid, alpha=0.2, color='tab:red')
plt.plot(range(6), mean_smi_final, c='green', marker='s', label='Epoch 50')
plt.fill_between(range(6), min_smi_final, max_smi_final, alpha=0.2, color='tab:green')

plt.xticks(range(6))
plt.xlabel(r'$T$', fontsize=14)
plt.ylabel(r'$SI (T;Y)$', fontsize=14)
plt.grid(True)
plt.legend(loc='upper left')
```

### Plots for Section 4.1: Memorization

#### 1. MLP, MNIST

#### a. 0 label noise

```python
cfg = {}
cfg['dataset'] = 'MNIST'
cfg['model'] = 'MLP'
cfg['width'] = [1024,1024,1024,1024]
cfg['optimizer'] = 'SGD'
cfg['noise_ratio'] = 0

run = 1
arch =  '-'.join(map(str,cfg['width']))
exp_name = 'label_noise/'+cfg['model']+'_'+arch+'_'+cfg['dataset']+'_'+cfg['optimizer']
if cfg['noise_ratio'] > 0:
    exp_name += '_LabelNoise_'+str(cfg['noise_ratio'])
    
cfg['exp_name'] = exp_name + '/run_%d'%(run)

with open(cfg['exp_name'] + '/metrics', 'rb') as f:
    metrics = pickle.load(f)
    
epochs_list = [0,2,4,6,8,10,20,30,40,50]
```

```python
with open(cfg['exp_name'] + '/metrics', 'rb') as f:
    metrics = pickle.load(f)

fig, ax1 = plt.subplots(figsize=(4,3), dpi=100)
ax2 = ax1.twinx()
ax1.plot(epochs_list, metrics['train_loss'], c='b', label='train loss')
ax1.plot(epochs_list, metrics['test_loss'], c='b', ls='--', label='test loss')
ax2.plot(epochs_list, metrics['train_acc'], c='r', label='train accuracy')
ax2.plot(epochs_list, metrics['test_acc'], c='r', ls='--', label='test accuracy')
ax1.grid(True)
ax2.spines['left'].set_color('blue')
ax2.spines['right'].set_color('red')
ax1.tick_params(axis='y', colors='blue')
ax2.tick_params(axis='y', colors='red')
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax2.set_ylabel('Accuracy', fontsize=12)
ax1.yaxis.label.set_color('b')
ax2.yaxis.label.set_color('r')
ax1.set_ylim(-0.01,2.5)
ax2.set_ylim(0,1.01)

# h1, l1 = ax1.get_legend_handles_labels()
# h2, l2 = ax2.get_legend_handles_labels()
# ax1.legend(h1+h2, l1+l2, loc='center right')
```

```python
with open(cfg['exp_name'] + '/smi', 'rb') as f:
    smi = pickle.load(f)

smi_all = defaultdict(list)

plt.figure(figsize=(4,3),dpi=100)
for i in range(4):
    for epoch in epochs_list:
        smi_all['Layer%d'%(i+1)].append(smi['Epoch%02d'%epoch][i])
    plt.plot(epochs_list, smi_all['Layer%d'%(i+1)], label='Layer %d'%(i+1), marker='o', ms=4)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel(r'$SI (T;Y)$', fontsize=14)
plt.ylim(0,1.3)
plt.grid(True)
```
